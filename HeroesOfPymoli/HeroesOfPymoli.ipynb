```python
import os
import pandas as pd

#Files to load
file_to_load= "Resources/purchase_data.csv"

#read file and store in panda df
purchase_data_df = pd.read_csv(file_to_load)
```

```python
purchase_data_df
```

```python
total_players=purchase_data_df["SN"].nunique()

total_players_df= pd.DataFrame({"Total Players": [total_players]})

total_players_df

```

```python
#calculate number of unique items
number_of_unique= purchase_data_df["Item Name"].nunique()

purchase_data_df["Number of Unique Items"]= number_of_unique

print(number_of_unique)

#calculate average price
average_price_number=purchase_data_df["Price"].mean()

purchase_data_df["Average Price of Item"]=average_price_number

print(average_price_number)

#calculate total number of purchases
number_of_purchases=purchase_data_df["Purchase ID"].count()

print(number_of_purchases)

total_revenue=purchase_data_df["Price"].sum()
print(total_revenue)

#create summary table of this info

summary_df=pd.DataFrame({ "Number of Unique Items" : [number_of_unique],
                        "Average Price" : [average_price_number],
                        "Number of Purchases" : [number_of_purchases],
                        "Total Revenue" : [total_revenue]})

summary_df.round(decimals=2)

summary_df
```

```python
#calculate total number of males and females

gender_group=purchase_data_df.groupby('Gender', as_index=True)

gender_count=gender_group["SN"].nunique()

gender_count

#calculate percentage of each gender from total player pool
gender_percentage=(gender_count/gender_count.sum()* 100)

gender_percentage

#display information in a table
gender_summary=pd.DataFrame({"Total Count": gender_count,
                            "Percentage of Players": gender_percentage})


gender_summary

```

```python
#find purchase data based on gender

gender_group = purchase_data_df.groupby("Gender")
gender_purchase= gender_group["Purchase ID"].count()
avg_price = gender_group["Price"].mean()
total_price= gender_group["Price"].sum()
avg_per_person=gender_group["Price"].sum()/gender_purchase

gender_purchase_summary=pd.DataFrame({"Purchase Count": gender_purchase,
                                     "Average Purchase Price" : avg_price,
                                     "Total Purchase Value" : total_price,
                                     "Average Total Purchase per Person" : avg_per_person})
gender_purchase_summary

```

```python
#creat bins for age groups and place players in appropriate bins
bins=[0, 9.5, 14.5, 19.5, 24.5, 29.5, 34.5, 39.5, 200]
names= ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#place existing players in appropriate age bins
purchase_data_df["Age Group"]= pd.cut(purchase_data_df["Age"], bins, labels=names)
purchase_data_df


age_group=purchase_data_df.groupby("Age Group")

#get count and percentage of ages
player_count= age_group["SN"].nunique()
age_percentage= player_count/total_players * 100

age_demographics=pd.DataFrame({"Total count": player_count,
                              "Percentage of Players": age_percentage})

age_demographics
```
